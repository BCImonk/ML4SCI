```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, precision_score, recall_score, roc_auc_score, cohen_kappa_score
from scipy.stats import kurtosis, skew
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap
import time

torch.manual_seed(1234)
np.random.seed(1234)
if torch.cuda.is_available():
    torch.cuda.manual_seed(1234)
    torch.backends.cudnn.deterministic = True

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

class EEGDataset(Dataset):
    def __init__(self, features, targets):
        self.features = features
        self.targets = targets

    def __len__(self):
        return len(self.targets)

    def __getitem__(self, idx):
        return self.features[idx], self.targets[idx]

class SVMInspiredModel(nn.Module):
    def __init__(self, input_dim):
        super(SVMInspiredModel, self).__init__()
        self.linear = nn.Linear(input_dim, 1)

    def forward(self, x):
        return self.linear(x).squeeze()

class EEGClassifier(nn.Module):
    def __init__(self, input_dim, dropout_rate=0.5):
        super(EEGClassifier, self).__init__()

        self.model = nn.Sequential(
            nn.Linear(input_dim, 32),
            nn.ReLU(),
            nn.BatchNorm1d(32),
            nn.Dropout(dropout_rate),

            nn.Linear(32, 16),
            nn.ReLU(),
            nn.BatchNorm1d(16),
            nn.Dropout(dropout_rate),

            nn.Linear(16, 8),
            nn.ReLU(),
            nn.BatchNorm1d(8),
            nn.Dropout(dropout_rate/2),

            nn.Linear(8, 1)
        )

        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.kaiming_normal_(m.weight)
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)

    def forward(self, x):
        return self.model(x).squeeze()

def extract_eeg_features(X, feature_names):
    band_indices = {
        'delta': [i for i, name in enumerate(feature_names) if 'delta' in name],
        'theta': [i for i, name in enumerate(feature_names) if 'theta' in name],
        'alpha': [i for i, name in enumerate(feature_names) if 'alpha' in name],
        'beta': [i for i, name in enumerate(feature_names) if 'beta' in name],
        'gamma': [i for i, name in enumerate(feature_names) if 'gamma' in name]
    }

    delta_indices = band_indices['delta']
    X_delta = X[:, delta_indices]

    top_channels = [23, 52, 45, 47, 17, 48, 10, 46, 55, 49]

    top_delta_indices = [delta_indices[i-1] for i in top_channels if i-1 < len(delta_indices)]
    X_top_delta = X[:, top_delta_indices]

    delta_mean = np.mean(X_delta, axis=1, keepdims=True)
    delta_var = np.var(X_delta, axis=1, keepdims=True)
    delta_max = np.max(X_delta, axis=1, keepdims=True)
    delta_min = np.min(X_delta, axis=1, keepdims=True)
    delta_range = delta_max - delta_min
    delta_kurt = kurtosis(X_delta, axis=1, keepdims=True)
    delta_skew = skew(X_delta, axis=1, keepdims=True)

    combined_features = [X_top_delta, delta_mean, delta_var, delta_max, delta_min, delta_range, delta_kurt, delta_skew]

    if band_indices['theta'] and band_indices['alpha'] and band_indices['beta']:
        theta_mean = np.mean(X[:, band_indices['theta']], axis=1, keepdims=True)
        alpha_mean = np.mean(X[:, band_indices['alpha']], axis=1, keepdims=True)
        beta_mean = np.mean(X[:, band_indices['beta']], axis=1, keepdims=True)

        delta_theta_ratio = delta_mean / (theta_mean + 1e-10)
        delta_alpha_ratio = delta_mean / (alpha_mean + 1e-10)
        delta_beta_ratio = delta_mean / (beta_mean + 1e-10)

        combined_features.extend([delta_theta_ratio, delta_alpha_ratio, delta_beta_ratio])

    X_features = np.hstack(combined_features)

    return X_features, band_indices

def augment_data(X, y, n_samples=500):
    X_aug = X.copy()
    y_aug = y.copy()

    X_class0 = X[y == 0]
    X_class1 = X[y == 1]

    n_class0 = len(X_class0)
    n_class1 = len(X_class1)

    n_samples_per_class = n_samples // 2

    for _ in range(n_samples_per_class):
        aug_strategy = np.random.randint(0, 5)

        if aug_strategy == 0:
            idx = np.random.randint(0, n_class0)
            base_sample = X_class0[idx]
            noise_level = np.random.uniform(0.01, 0.05)
            noise = np.random.normal(0, noise_level, base_sample.shape)
            new_sample = base_sample + noise

        elif aug_strategy == 1:
            idx1, idx2 = np.random.randint(0, n_class0, 2)
            alpha = np.random.beta(0.4, 0.4)
            new_sample = alpha * X_class0[idx1] + (1 - alpha) * X_class0[idx2]

        elif aug_strategy == 2:
            idx1, idx2 = np.random.randint(0, n_class0, 2)
            alpha = np.random.beta(0.4, 0.4)
            interpolated = alpha * X_class0[idx1] + (1 - alpha) * X_class0[idx2]
            noise_level = np.random.uniform(0.01, 0.03)
            noise = np.random.normal(0, noise_level, interpolated.shape)
            new_sample = interpolated + noise

        elif aug_strategy == 3:
            idx = np.random.randint(0, n_class0)
            scale = np.random.uniform(0.9, 1.1)
            new_sample = X_class0[idx] * scale

        else:
            idx = np.random.randint(0, n_class0)
            base_sample = X_class0[idx].copy()
            features_to_jitter = np.random.choice(
                range(len(base_sample)),
                size=int(0.3 * len(base_sample)),
                replace=False
            )
            for feat_idx in features_to_jitter:
                base_sample[feat_idx] *= np.random.uniform(0.85, 1.15)
            new_sample = base_sample

        X_aug = np.vstack([X_aug, new_sample.reshape(1, -1)])
        y_aug = np.append(y_aug, 0)

    for _ in range(n_samples_per_class):
        aug_strategy = np.random.randint(0, 5)

        if aug_strategy == 0:
            idx = np.random.randint(0, n_class1)
            base_sample = X_class1[idx]
            noise_level = np.random.uniform(0.01, 0.05)
            noise = np.random.normal(0, noise_level, base_sample.shape)
            new_sample = base_sample + noise

        elif aug_strategy == 1:
            idx1, idx2 = np.random.randint(0, n_class1, 2)
            alpha = np.random.beta(0.4, 0.4)
            new_sample = alpha * X_class1[idx1] + (1 - alpha) * X_class1[idx2]

        elif aug_strategy == 2:
            idx1, idx2 = np.random.randint(0, n_class1, 2)
            alpha = np.random.beta(0.4, 0.4)
            interpolated = alpha * X_class1[idx1] + (1 - alpha) * X_class1[idx2]
            noise_level = np.random.uniform(0.01, 0.03)
            noise = np.random.normal(0, noise_level, interpolated.shape)
            new_sample = interpolated + noise

        elif aug_strategy == 3:
            idx = np.random.randint(0, n_class1)
            scale = np.random.uniform(0.9, 1.1)
            new_sample = X_class1[idx] * scale

        else:
            idx = np.random.randint(0, n_class1)
            base_sample = X_class1[idx].copy()
            features_to_jitter = np.random.choice(
                range(len(base_sample)),
                size=int(0.3 * len(base_sample)),
                replace=False
            )
            for feat_idx in features_to_jitter:
                base_sample[feat_idx] *= np.random.uniform(0.85, 1.15)
            new_sample = base_sample

        X_aug = np.vstack([X_aug, new_sample.reshape(1, -1)])
        y_aug = np.append(y_aug, 1)

    return X_aug, y_aug

def hinge_loss(outputs, targets):
    targets = 2 * targets - 1
    losses = torch.clamp(1 - outputs * targets, min=0)
    return torch.mean(losses)

def train_model(model, train_loader, val_loader, criterion, optimizer, device, n_epochs=200, patience=20):
    best_val_acc = 0
    best_model_state = None
    patience_counter = 0

    history = {
        'train_loss': [],
        'val_accuracy': [],
        'val_f1': [],
        'val_auc': [],
        'val_kappa': []
    }

    for epoch in range(n_epochs):
        model.train()
        train_loss = 0

        for inputs, targets in train_loader:
            inputs, targets = inputs.to(device), targets.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, targets)

            loss.backward()
            optimizer.step()

            train_loss += loss.item()

        model.eval()
        val_preds = []
        val_targets = []
        val_probs = []

        with torch.no_grad():
            for inputs, targets in val_loader:
                inputs, targets = inputs.to(device), targets.to(device)
                outputs = model(inputs)

                if isinstance(criterion, nn.BCEWithLogitsLoss):
                    probs = torch.sigmoid(outputs).cpu().numpy()
                    preds = (outputs > 0).float().cpu().numpy()
                else:
                    probs = (torch.sigmoid(outputs * 0.5 + 0.5)).cpu().numpy()
                    preds = (outputs > 0).float().cpu().numpy()

                val_probs.extend(probs)
                val_preds.extend(preds)
                val_targets.extend(targets.cpu().numpy())

        val_acc = accuracy_score(val_targets, val_preds)
        val_f1 = f1_score(val_targets, val_preds)
        val_auc = roc_auc_score(val_targets, val_probs) if len(np.unique(val_targets)) > 1 else 0.5
        val_kappa = cohen_kappa_score(val_targets, val_preds)

        history['train_loss'].append(train_loss/len(train_loader))
        history['val_accuracy'].append(val_acc)
        history['val_f1'].append(val_f1)
        history['val_auc'].append(val_auc)
        history['val_kappa'].append(val_kappa)

        if (epoch + 1) % 10 == 0:
            print(f"Epoch {epoch+1}, Train Loss: {train_loss/len(train_loader):.4f}, Val Acc: {val_acc:.4f}, F1: {val_f1:.4f}, AUC: {val_auc:.4f}")

        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_model_state = model.state_dict().copy()
            patience_counter = 0
        else:
            patience_counter += 1
            if patience_counter >= patience and epoch > 30:
                print(f"Early stopping at epoch {epoch+1}")
                break

    if best_model_state:
        model.load_state_dict(best_model_state)

    return model, best_val_acc, val_probs, val_targets, history

def plot_training_history(history, title):
    plt.figure(figsize=(15, 6))

    plt.subplot(1, 2, 1)
    plt.plot(history['train_loss'], label='Train Loss')
    plt.title('Training Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history['val_accuracy'], label='Accuracy')
    plt.plot(history['val_f1'], label='F1 Score')
    plt.plot(history['val_auc'], label='AUC')
    plt.plot(history['val_kappa'], label='Kappa')
    plt.title('Validation Metrics')
    plt.xlabel('Epoch')
    plt.ylabel('Score')
    plt.legend()

    plt.suptitle(title)
    plt.tight_layout()
    plt.show()

def plot_confusion_matrix(cm, title='Confusion Matrix'):
    plt.figure(figsize=(8, 6))
    is_integer = np.all(np.equal(np.mod(cm, 1), 0))
    fmt = 'd' if is_integer else '.2f'
    sns.heatmap(cm, annot=True, fmt=fmt, cmap="Blues")
    plt.title(title)
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.show()

def visualize_feature_importances(model, feature_names, title='Feature Importances'):
    if isinstance(model, SVMInspiredModel):
        weights = model.linear.weight.cpu().detach().numpy()[0]
    else:
        weights = model.model[0].weight.cpu().detach().numpy().mean(axis=0)

    abs_weights = np.abs(weights)
    importance_indices = np.argsort(abs_weights)[::-1]

    plt.figure(figsize=(10, 8))
    plt.barh(range(len(feature_names)), abs_weights[importance_indices])
    plt.yticks(range(len(feature_names)), [feature_names[i] for i in importance_indices])
    plt.title(title)
    plt.xlabel('Absolute Weight')
    plt.tight_layout()
    plt.show()

def plot_eeg_channel_importance(model, band_indices, top_channels):
    if isinstance(model, SVMInspiredModel):
        weights = model.linear.weight.cpu().detach().numpy()[0]
    else:
        weights = model.model[0].weight.cpu().detach().numpy().mean(axis=0)

    delta_indices = band_indices['delta']
    delta_weights = np.zeros(64)

    for i, idx in enumerate(delta_indices):
        if idx < len(weights):
            delta_weights[i] = abs(weights[idx])

    plt.figure(figsize=(10, 8))
    plt.scatter(range(1, 65), delta_weights, alpha=0.5)
    plt.scatter([ch for ch in top_channels if ch <= 64],
                [delta_weights[ch-1] for ch in top_channels if ch <= 64],
                color='r', label='Top Channels')

    plt.title('EEG Delta Channel Importance')
    plt.xlabel('Channel Number')
    plt.ylabel('Absolute Weight')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

def plot_band_distributions(X, band_indices, class_labels):
    bands = ['delta', 'theta', 'alpha', 'beta', 'gamma']
    plt.figure(figsize=(15, 10))

    for i, band in enumerate(bands):
        if band_indices[band]:
            band_data = X[:, band_indices[band]]
            band_mean = np.mean(band_data, axis=1)

            plt.subplot(2, 3, i+1)
            for label in np.unique(class_labels):
                sns.kdeplot(band_mean[class_labels == label], label=f'Class {label}')

            plt.title(f'{band.capitalize()} Band Power Distribution')
            plt.xlabel('Mean Power')
            plt.ylabel('Density')
            plt.legend()

    plt.tight_layout()
    plt.show()

def visualize_augmented_data(X_orig, y_orig, X_aug, y_aug):
    from sklearn.decomposition import PCA

    pca = PCA(n_components=2)
    X_orig_pca = pca.fit_transform(X_orig)
    X_aug_pca = pca.transform(X_aug[len(X_orig):])

    plt.figure(figsize=(10, 8))

    plt.scatter(X_orig_pca[y_orig==0, 0], X_orig_pca[y_orig==0, 1],
                c='blue', label='Original Class 0', marker='o', alpha=0.8)
    plt.scatter(X_orig_pca[y_orig==1, 0], X_orig_pca[y_orig==1, 1],
                c='red', label='Original Class 1', marker='o', alpha=0.8)

    plt.scatter(X_aug_pca[y_aug[len(X_orig):]==0, 0], X_aug_pca[y_aug[len(X_orig):]==0, 1],
                c='skyblue', label='Augmented Class 0', marker='x', alpha=0.5)
    plt.scatter(X_aug_pca[y_aug[len(X_orig):]==1, 0], X_aug_pca[y_aug[len(X_orig):]==1, 1],
                c='pink', label='Augmented Class 1', marker='x', alpha=0.5)

    plt.title('PCA of Original vs Augmented Data')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

def calculate_metrics(y_true, y_pred, y_prob=None):
    metrics = {
        'accuracy': accuracy_score(y_true, y_pred),
        'precision': precision_score(y_true, y_pred),
        'recall': recall_score(y_true, y_pred),
        'f1': f1_score(y_true, y_pred),
        'specificity': (y_pred[y_true==0] == 0).sum() / (y_true==0).sum(),
        'sensitivity': recall_score(y_true, y_pred),
        'kappa': cohen_kappa_score(y_true, y_pred)
    }

    if y_prob is not None:
        metrics['auc'] = roc_auc_score(y_true, y_prob)

    return metrics

def create_feature_names(band_indices, top_channels):
    feature_names = []

    for i, channel in enumerate(top_channels):
        feature_names.append(f"delta_{channel}")

    feature_names.extend([
        "delta_mean", "delta_var", "delta_max", "delta_min",
        "delta_range", "delta_kurt", "delta_skew",
        "delta_theta_ratio", "delta_alpha_ratio", "delta_beta_ratio"
    ])

    return feature_names

def train_and_evaluate():
    print("Loading data...")
    df = pd.read_csv('/eeg data.csv')

    X = df.iloc[:, 1:-1].values
    y = df.iloc[:, -1].values.astype(np.float32)
    feature_names = df.columns[1:-1].tolist()

    print("\nDataset Overview:")
    print(f"Total samples: {len(X)}")
    print(f"Class distribution: {np.bincount(y.astype(int))}")
    print(f"Original features: {X.shape[1]}")

    plt.figure(figsize=(8, 5))
    plt.bar(['Class 0', 'Class 1'], np.bincount(y.astype(int)))
    plt.title('Class Distribution')
    plt.ylabel('Count')
    plt.grid(axis='y', alpha=0.3)
    plt.show()

    print("\nExtracting features...")
    X_features, band_indices = extract_eeg_features(X, feature_names)

    print("\nBand statistics:")
    for band, indices in band_indices.items():
        if indices:
            band_data = X[:, indices]
            print(f"{band.capitalize()} band: mean={np.mean(band_data):.4f}, std={np.std(band_data):.4f}, min={np.min(band_data):.4f}, max={np.max(band_data):.4f}")

    plot_band_distributions(X, band_indices, y)

    top_channels = [23, 52, 45, 47, 17, 48, 10, 46, 55, 49]
    custom_feature_names = create_feature_names(band_indices, top_channels)

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_features)

    print(f"\nUsing {X_scaled.shape[1]} engineered features")

    n_folds = 5
    kf = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=1234)

    all_fold_probs = np.zeros((len(X_scaled), n_folds, 2))
    oof_indices = []

    fold_results = []
    fold_histories = []
    fold_models = []

    print(f"\nTraining with {n_folds}-fold cross-validation...")

    for fold, (train_idx, val_idx) in enumerate(kf.split(X_scaled, y)):
        print(f"\nFold {fold+1}/{n_folds}")

        X_train, X_val = X_scaled[train_idx], X_scaled[val_idx]
        y_train, y_val = y[train_idx], y[val_idx]

        oof_indices.append(val_idx)

        print("Augmenting training data...")
        X_train_aug, y_train_aug = augment_data(X_train, y_train, n_samples=500)
        print(f"Training samples: {len(X_train)}  {len(X_train_aug)}")

        if fold == 0:
            visualize_augmented_data(X_train, y_train, X_train_aug, y_train_aug)

        train_dataset = EEGDataset(torch.FloatTensor(X_train_aug), torch.FloatTensor(y_train_aug))
        val_dataset = EEGDataset(torch.FloatTensor(X_val), torch.FloatTensor(y_val))

        train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
        val_loader = DataLoader(val_dataset, batch_size=32)

        print("Training SVM-inspired model...")
        svm_model = SVMInspiredModel(input_dim=X_train.shape[1]).to(device)
        svm_criterion = hinge_loss
        svm_optimizer = optim.SGD(svm_model.parameters(), lr=0.01, weight_decay=0.01)

        trained_svm, svm_acc, svm_probs, svm_targets, svm_history = train_model(
            svm_model, train_loader, val_loader,
            svm_criterion, svm_optimizer, device,
            n_epochs=200, patience=30
        )

        print("Training neural network model...")
        nn_model = EEGClassifier(input_dim=X_train.shape[1], dropout_rate=0.5).to(device)
        nn_criterion = nn.BCEWithLogitsLoss()
        nn_optimizer = optim.Adam(nn_model.parameters(), lr=0.001, weight_decay=0.01)

        trained_nn, nn_acc, nn_probs, nn_targets, nn_history = train_model(
            nn_model, train_loader, val_loader,
            nn_criterion, nn_optimizer, device,
            n_epochs=200, patience=30
        )

        plot_training_history(svm_history, f"Fold {fold+1} - SVM Model Training")
        plot_training_history(nn_history, f"Fold {fold+1} - Neural Network Training")

        all_fold_probs[val_idx, fold, 0] = np.array(svm_probs)
        all_fold_probs[val_idx, fold, 1] = np.array(nn_probs)

        if svm_acc >= nn_acc:
            print(f"SVM model performed better (Acc: {svm_acc:.4f} vs {nn_acc:.4f})")
            best_model = trained_svm
            best_probs = svm_probs
            best_history = svm_history
            model_type = "svm"
        else:
            print(f"Neural network model performed better (Acc: {nn_acc:.4f} vs {svm_acc:.4f})")
            best_model = trained_nn
            best_probs = nn_probs
            best_history = nn_history
            model_type = "nn"

        fold_models.append((best_model, model_type))
        fold_histories.append(best_history)

        best_threshold = 0.5
        best_threshold_acc = accuracy_score(y_val, (np.array(best_probs) > 0.5).astype(float))

        for threshold in np.arange(0.3, 0.7, 0.02):
            threshold_preds = (np.array(best_probs) > threshold).astype(float)
            threshold_acc = accuracy_score(y_val, threshold_preds)

            if threshold_acc > best_threshold_acc:
                best_threshold_acc = threshold_acc
                best_threshold = threshold

        best_preds = (np.array(best_probs) > best_threshold).astype(float)
        metrics = calculate_metrics(y_val, best_preds, best_probs)

        print(f"\nFold {fold+1} Results (threshold={best_threshold:.2f}):")
        for metric, value in metrics.items():
            print(f"{metric}: {value:.4f}")

        cm = confusion_matrix(y_val, best_preds)
        plot_confusion_matrix(cm, f"Fold {fold+1} Confusion Matrix")

        if fold == 0:
            if model_type == "svm":
                visualize_feature_importances(best_model, custom_feature_names, "SVM Feature Importances")
                plot_eeg_channel_importance(best_model, band_indices, top_channels)
            else:
                visualize_feature_importances(best_model, custom_feature_names, "Neural Network Feature Importances")

        fold_results.append({
            'fold': fold + 1,
            'model_type': model_type,
            'threshold': best_threshold,
            'metrics': metrics
        })

        torch.save(trained_svm.state_dict(), f'svm_model_fold{fold+1}.pt')
        torch.save(trained_nn.state_dict(), f'nn_model_fold{fold+1}.pt')

    print("\nCross-validation results:")
    cv_acc = [result['metrics']['accuracy'] for result in fold_results]
    cv_f1 = [result['metrics']['f1'] for result in fold_results]
    cv_auc = [result['metrics']['auc'] for result in fold_results if 'auc' in result['metrics']]
    cv_thresholds = [result['threshold'] for result in fold_results]

    print(f"Mean Accuracy: {np.mean(cv_acc):.4f}  {np.std(cv_acc):.4f}")
    print(f"Mean F1 Score: {np.mean(cv_f1):.4f}  {np.std(cv_f1):.4f}")
    if cv_auc:
        print(f"Mean AUC: {np.mean(cv_auc):.4f}  {np.std(cv_auc):.4f}")

    plt.figure(figsize=(12, 6))

    metrics_to_plot = ['accuracy', 'precision', 'recall', 'f1', 'specificity', 'sensitivity', 'kappa']
    if 'auc' in fold_results[0]['metrics']:
        metrics_to_plot.append('auc')

    fold_numbers = [result['fold'] for result in fold_results]

    for metric in metrics_to_plot:
        metric_values = [result['metrics'][metric] for result in fold_results]
        plt.plot(fold_numbers, metric_values, marker='o', label=metric)

    plt.title('Model Performance Across Folds')
    plt.xlabel('Fold')
    plt.ylabel('Score')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

    print("\nCreating ensemble predictions...")

    ensemble_probs = np.zeros(len(X_scaled))

    for i in range(len(X_scaled)):
        fold_probs = []

        for fold, val_idx in enumerate(oof_indices):
            if i in val_idx:
                fold_probs.append((all_fold_probs[i, fold, 0] + all_fold_probs[i, fold, 1]) / 2)

        if fold_probs:
            ensemble_probs[i] = np.mean(fold_probs)
        else:
            ensemble_probs[i] = 0.5

    best_ens_threshold = np.mean(cv_thresholds)
    best_ens_acc = 0

    for threshold in np.arange(0.3, 0.7, 0.01):
        ens_preds = (ensemble_probs > threshold).astype(float)
        ens_acc = accuracy_score(y, ens_preds)

        if ens_acc > best_ens_acc:
            best_ens_acc = ens_acc
            best_ens_threshold = threshold

    ensemble_preds = (ensemble_probs > best_ens_threshold).astype(float)
    ensemble_metrics = calculate_metrics(y, ensemble_preds, ensemble_probs)
    ensemble_cm = confusion_matrix(y, ensemble_preds)

    print(f"\nEnsemble results with threshold {best_ens_threshold:.2f}:")
    for metric, value in ensemble_metrics.items():
        print(f"{metric}: {value:.4f}")

    plot_confusion_matrix(ensemble_cm, "Ensemble Model Confusion Matrix")

    print("\nTraining final models on all data...")

    X_full_aug, y_full_aug = augment_data(X_scaled, y, n_samples=600)

    full_dataset = EEGDataset(torch.FloatTensor(X_full_aug), torch.FloatTensor(y_full_aug))
    full_loader = DataLoader(full_dataset, batch_size=32, shuffle=True)

    final_svm = SVMInspiredModel(input_dim=X_scaled.shape[1]).to(device)
    svm_criterion = hinge_loss
    svm_optimizer = optim.SGD(final_svm.parameters(), lr=0.01, weight_decay=0.01)

    final_nn = EEGClassifier(input_dim=X_scaled.shape[1], dropout_rate=0.5).to(device)
    nn_criterion = nn.BCEWithLogitsLoss()
    nn_optimizer = optim.Adam(final_nn.parameters(), lr=0.001, weight_decay=0.01)

    svm_history = {'train_loss': []}
    nn_history = {'train_loss': []}

    for epoch in range(150):
        final_svm.train()
        svm_loss = 0

        for inputs, targets in full_loader:
            inputs, targets = inputs.to(device), targets.to(device)

            svm_optimizer.zero_grad()
            outputs = final_svm(inputs)
            loss = svm_criterion(outputs, targets)

            loss.backward()
            svm_optimizer.step()

            svm_loss += loss.item()

        final_nn.train()
        nn_loss = 0

        for inputs, targets in full_loader:
            inputs, targets = inputs.to(device), targets.to(device)

            nn_optimizer.zero_grad()
            outputs = final_nn(inputs)
            loss = nn_criterion(outputs, targets)

            loss.backward()
            nn_optimizer.step()

            nn_loss += loss.item()

        svm_history['train_loss'].append(svm_loss/len(full_loader))
        nn_history['train_loss'].append(nn_loss/len(full_loader))

        if (epoch + 1) % 20 == 0:
            print(f"Epoch {epoch+1}, SVM Loss: {svm_loss/len(full_loader):.4f}, NN Loss: {nn_loss/len(full_loader):.4f}")

    plt.figure(figsize=(12, 5))
    plt.plot(svm_history['train_loss'], label='SVM Loss')
    plt.plot(nn_history['train_loss'], label='NN Loss')
    plt.title('Final Model Training Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

    torch.save(final_svm.state_dict(), 'final_svm_model.pt')
    torch.save(final_nn.state_dict(), 'final_nn_model.pt')

    final_dataset = EEGDataset(torch.FloatTensor(X_scaled), torch.FloatTensor(y))
    final_loader = DataLoader(final_dataset, batch_size=32)

    final_svm.eval()
    svm_probs = []

    with torch.no_grad():
        for inputs, _ in final_loader:
            inputs = inputs.to(device)
            outputs = final_svm(inputs)
            probs = (torch.sigmoid(outputs * 0.5 + 0.5)).cpu().numpy()
            svm_probs.extend(probs)

    svm_preds = (np.array(svm_probs) > best_ens_threshold).astype(float)
    svm_metrics = calculate_metrics(y, svm_preds, svm_probs)
    svm_cm = confusion_matrix(y, svm_preds)

    final_nn.eval()
    nn_probs = []

    with torch.no_grad():
        for inputs, _ in final_loader:
            inputs = inputs.to(device)
            outputs = final_nn(inputs)
            probs = torch.sigmoid(outputs).cpu().numpy()
            nn_probs.extend(probs)

    nn_preds = (np.array(nn_probs) > best_ens_threshold).astype(float)
    nn_metrics = calculate_metrics(y, nn_preds, nn_probs)
    nn_cm = confusion_matrix(y, nn_preds)

    combined_probs = [(svm_probs[i] + nn_probs[i])/2 for i in range(len(svm_probs))]
    combined_preds = (np.array(combined_probs) > best_ens_threshold).astype(float)
    combined_metrics = calculate_metrics(y, combined_preds, combined_probs)
    combined_cm = confusion_matrix(y, combined_preds)

    print("\nFinal SVM model performance:")
    for metric, value in svm_metrics.items():
        print(f"{metric}: {value:.4f}")
    plot_confusion_matrix(svm_cm, "Final SVM Model Confusion Matrix")

    print("\nFinal Neural Network model performance:")
    for metric, value in nn_metrics.items():
        print(f"{metric}: {value:.4f}")
    plot_confusion_matrix(nn_cm, "Final Neural Network Confusion Matrix")

    print("\nFinal Combined model performance:")
    for metric, value in combined_metrics.items():
        print(f"{metric}: {value:.4f}")
    plot_confusion_matrix(combined_cm, "Final Combined Model Confusion Matrix")

    all_accuracies = [ensemble_metrics['accuracy'], svm_metrics['accuracy'], nn_metrics['accuracy'], combined_metrics['accuracy']]
    all_f1_scores = [ensemble_metrics['f1'], svm_metrics['f1'], nn_metrics['f1'], combined_metrics['f1']]

    plt.figure(figsize=(12, 6))
    plt.bar(['Ensemble', 'SVM', 'Neural Network', 'Combined'], all_accuracies, alpha=0.7, label='Accuracy')
    plt.bar(['Ensemble', 'SVM', 'Neural Network', 'Combined'], all_f1_scores, alpha=0.7, label='F1 Score')
    plt.title('Final Model Comparison')
    plt.ylabel('Score')
    plt.ylim(0, 1)
    plt.legend()
    plt.grid(axis='y', alpha=0.3)
    plt.show()

    visualize_feature_importances(final_svm, custom_feature_names, "Final SVM Feature Importances")
    visualize_feature_importances(final_nn, custom_feature_names, "Final Neural Network Feature Importances")

    best_acc = max(ensemble_metrics['accuracy'], svm_metrics['accuracy'], nn_metrics['accuracy'], combined_metrics['accuracy'])
    best_f1 = max(ensemble_metrics['f1'], svm_metrics['f1'], nn_metrics['f1'], combined_metrics['f1'])

    print("\nBest model summary:")
    print(f"Best Accuracy: {best_acc:.4f}")
    print(f"Best F1 Score: {best_f1:.4f}")

    return best_acc, best_f1

if __name__ == "__main__":
    start_time = time.time()
    acc, f1 = train_and_evaluate()
    end_time = time.time()

    print(f"\nTotal execution time: {(end_time - start_time) / 60:.2f} minutes")
    print(f"Final Accuracy: {acc:.4f}")
    print(f"Final F1 Score: {f1:.4f}")
```
